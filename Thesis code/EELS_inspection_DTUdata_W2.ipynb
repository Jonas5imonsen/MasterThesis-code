# EELS inspection of RuCl3 and RuO2

## Curve fitting quantification
### Prosedure: 
1. Inspect relative thickness $ln(I_t/I_0)$ (if it is larger than 0.5 FRD is needed!!!)
2. Remove background from core-loss signal by fitting power spectrum
1. Deconvolve the low loss spectrum to remove plural scattering with Fourier ratio deconvolution (FRD) (must be done before Hartree-Slater since it assumes single scattering)

3. integrate the edge intesity of interest
4. Estimate of atoms per unit area by Hartree–Slater calculations, $N_i = \frac{I_i}{\sigma_i}$
5. Calcualte ratios $N_{Cl}/N_{Ru}$ with $N_O/N_{Ru}$

## 5.

### Make and fit model

```python
# Calculate Hartree-Slater inelastic scattering-crossection

#make model 
model = s_deconvolved.create_model(low_loss=s_ll, auto_background = False) #adding low loss signal makes deconvolution possible
model.components #if elements in model is wrong all fitting will be wrong !!
```

```python
model.multifit(kind = 'smart')
```

```python
model.plot()
```

```python
edges = ("Ru_M5" ,"Cl_L2,3","O_K",)
```

### Improve fitting

```python
model.components.Ru_M5.onset_energy.value = 273.7
model.components.Ru_M5.onset_energy.assign_current_value_to_all()
```

```python
model.components.Cl_L23.onset_energy.value = 202.8
model.components.Cl_L23.onset_energy.assign_current_value_to_all()
```

```python
model.enable_fine_structure()
```

```python
model.multifit( kind = 'smart')
```

```python
model.plot()
```

### Extract Hartree-Slater cross-sections

```python
EDGES = [model["Cl_L2,3"],model['Ru_M4']]
boundaries = [(200,240),(295,345)]
HS_cross_sections = []
E = s_deconvolved.axes_manager[0].axis


for edge, boundary in zip(EDGES, boundaries):

    A = edge.intensity.value

    '''# Turn off fine structure
    O_K.fine_structure_active = False'''
    I_E = edge.function(E)

    sigma_E = I_E / A

    mask = (E > boundary[0]) & (E < boundary[1])
    sigma = np.trapz(sigma_E[mask], E[mask])

    HS_cross_sections.append(sigma)
```

```python
print(HS_cross_sections)
```

```python
N_Cl = Cl_L_intensity/HS_cross_sections[0]
N_Ru = Ru_M4_intensity/HS_cross_sections[1]
```

```python
print(N_Cl)
print(N_Ru)
print(N_Cl/N_Ru)
```
