```python
%matplotlib qt5
import hyperspy.api as hs
import pyxem as pxm
import numpy as np
import matplotlib.pyplot as plt

from zarr import ZipStore
from pathlib import Path
from mpl_toolkits.axes_grid1.anchored_artists import AnchoredSizeBar
```

```python
datapath = Path(r"/Users/jonasfsimonsen/Documents/NTNU/5klasse/Masteroppgave/DataHandling/SPED/transfer_460455_files_78139f15/20260515_142854_SPED_128x128_5x5_1p745x1p745nm_12cm_NBD_a5_spot1p0nm_1deg_100Hz.zspy")
store = ZipStore(datapath)
signal = hs.load(store, lazy=True)
signal.set_signal_type('electron_diffraction')
```

```python
signal.original_metadata
```

```python
signal.axes_manager
```

## Calibrate navigation data from metadata

```python
signal.axes_manager[0].scale = float(signal.metadata.Metadata.metadata.dx)
signal.axes_manager[1].scale = float(signal.metadata.Metadata.metadata.dy)

signal.axes_manager[0].units='nm'
signal.axes_manager[1].units='nm'

signal.axes_manager
```

## Inspect signal by plotting

```python
signal.plot(norm='symlog', cmap='viridis')
```

```python
signal.T.plot(norm='symlog', cmap='viridis')
```

## Radial virtual DF

```python
signalT = signal.T
```

```python
nx, ny = signal.axes_manager.signal_shape
cx, cy = nx//2, ny//2
roi = hs.roi.CircleROI(cx, cy, 10) #position in pixels
```

```python
max_sig = signal.max(axis=[0, 1])
max_sig.compute() #Perform the calculations on the lazily loaded data
```

```python
signalT.plot(navigator=max_sig,norm='symlog', cmap='viridis')
```

```python
roi.add_widget(signalT)
```

```python
roi_params = {
    "cx": roi.cx,
    "cy": roi.cy,
    "r": roi.r,
    "r_inner": roi.r_inner
}
print(roi_params)
```

```python
# Recreate ROI
roi2 = hs.roi.CircleROI(
    cx=roi_params["cx"],
    cy=roi_params["cy"],
    r=roi_params["r"],
    r_inner=roi_params["r_inner"]
)
```

```python
# Plot virtual DF image
signal.plot_integrated_intensity(roi=roi2,cmap='viridis', norm='symlog')
#signal.compute()
```

### Save DF image and the virtual detector

```python
signal.plot_integrated_intensity(
    roi=roi2,
    cmap='magma',
    norm='symlog',
    colorbar=None,
    axes_off=True
)

fig = plt.gcf()
ax = plt.gca()

# Remove title
ax.set_title("")

plt.tight_layout(pad=0)

fig.savefig(
    'test.pdf',
    dpi=500,
    bbox_inches='tight',
    pad_inches=0
)
```

```python
virtualDetector = roi2(signal).sum(axis=(0, 1))
```

```python
fig, ax = plt.subplots()
ax.imshow(virtualDetector.data, norm = 'symlog', cmap='viridis')
ax.set_xticks([])
ax.set_yticks([])
fig.savefig('R1_virtualDetector_sum_vir.pdf', dpi = 350)
```

```python
signal.sum().plot(vmin=0,vmax=400000, cmap='gist_gray',axes_off=True,colorbar=None)
fig = plt.gcf()
ax = plt.gca()

# Remove title
ax.set_title("")

plt.tight_layout(pad=0)

fig.savefig(
    'sum_sig.pdf',
    dpi=1200,
    bbox_inches='tight',
    pad_inches=0
)
```
